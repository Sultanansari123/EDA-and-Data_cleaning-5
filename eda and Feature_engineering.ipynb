# Training and evaluating the accuracy of the model

```python
# importing the library
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
# Create instances of the classifiers
logreg = LogisticRegression()
rf = RandomForestClassifier()
svm = SVC()
```

```python
# Perform cross-validation and calculate accuracy scores
scores_logreg = cross_val_score(logreg, X, y, cv=5)
scores_rf = cross_val_score(rf, X, y, cv=5)
scores_svm = cross_val_score(svm, X, y, cv=5)
```

```python
# Calculate the mean accuracy for each model
mean_accuracy_logreg = scores_logreg.mean()
mean_accuracy_rf = scores_rf.mean()
mean_accuracy_svm = scores_svm.mean()
```

```python
# Print the mean accuracy scores
print("Logistic Regression Accuracy:", mean_accuracy_logreg)
print("Random Forest Accuracy:", mean_accuracy_rf)
print("Support Vector Machines Accuracy:", mean_accuracy_svm)
```

```python
# Find the model with the highest accuracy
best_model = max(mean_accuracy_logreg, mean_accuracy_rf, mean_accuracy_svm)*100
```

```python
# Print the name of the best model
if best_model == mean_accuracy_logreg:
    print("The best model is Logistic Regression:",best_model,"%")
elif best_model == mean_accuracy_rf:
    print("The best model is Random Forest:",best_model,"%")
else:
    print("The best model is Support Vector Machines:",best_model,"%")
```
